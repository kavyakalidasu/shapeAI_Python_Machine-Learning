```python
from sklearn.model_selection import train_test_split
X = boston.drop('MEDV', axis=1)
Y = boston['MEDV']
X_tr, X_te, Y_tr, Y_te=train_test_split(X, Y, test_size=0.15, random_state=4)
print(X_tr.shape)
print(X_te.shape)
print(Y_tr.shape)
print(Y_te.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
#fitting model
lm=LinearRegression()
lm.fit(X_tr ,Y_tr)
#lm.fit(X_tr ,Y_tr)
```

```python
y_t_pr=lm.predict(X_tr)
rmse=np.sqrt(mean_squared_error(Y_tr, y_t_pr))
print("The model performance for training data")
print('RMSE is{}'.format(rmse))
print("\n")
#on testing data
t_tp=lm.predict(X_te)
rmsep=np.sqrt(mean_squared_error(Y_te,t_tp))
print("The model performance for testing data")
print('RMSE is{}'.format(rmsep))
```
